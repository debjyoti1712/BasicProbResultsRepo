**BERNOULLI'S DISTRIBUTION (DISCRETE DISTRIBUTION)**

```python
# 1. BERNOULLI'S DISTRIBUTION (DISCRETE DISTRIBUTION)

# Ref : https://en.wikipedia.org/wiki/Bernoulli_distribution
# Ref(Histogram) : https://datatofish.com/plot-histogram-python/

import numpy as np
import matplotlib.pyplot as plt

# Returns factorial of n 
def fact(n):  
    res = 1     
    for i in range(2, n+1): 
        res = res * i          
    return res 

trials = 70  # No of trials
p=0.1
k=np.random.choice([0, 1], size=trials)  # this is data
# print(k)
y = (p**k)*((1-p)**(1-k))
# print(y)

n, bins, patches = plt.hist(k, 10, density=True) # hist , data per bin = 100 + 1

plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("Frequency --->", fontweight='bold') 
titleText = "Bernoulli's Distribution with p = " + str(p) + " ,Trials = " + str(trials) + "\n" 
plt.title(titleText, fontweight ="bold") 
plt.show()
```

**BINOMIAL DISTRIBUTION (DISCRETE DISTRIBUTION)**

```python
# 2. BINOMIAL DISTRIBUTION (DISCRETE DISTRIBUTION)

# Ref 1: https://www.mathsisfun.com/data/binomial-distribution.html
# Ref 2: https://www.itl.nist.gov/div898/handbook/eda/section3/eda366i.htm

import matplotlib.pyplot as plt

def nCr(n, r): 
  return (fact(n)/(fact(r)* fact(n - r))) 
  
# Returns factorial of n 
def fact(n):  
    res = 1     
    for i in range(2, n+1): 
        res = res * i          
    return res 
n=80  # No of trials
p=1/2
result=[]
for i in range(1,n+1):
  x=nCr(n,i)*(p**i)*((1-p)**(n-i))
  result.append(x)
  #print(x)
#plt.plot(range(0,n+1),result,"or",range(0,n+1),result)
plt.plot(result,"or",result)
titleText = "Binomial Distribution with p = " + str(p) + " , Trials = " + str(n) + "\n" 
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) = P(X=x) --->", fontweight='bold') 
plt.title(titleText, fontweight ="bold") 
plt.show()
```

**UNIFORM DISTRIBUTION(CONTINUOUS DISTRIBUTION)**

```python
# 3.  UNIFORM DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref: https://www.datacamp.com/community/tutorials/probability-distributions-python

import numpy as np
import matplotlib.pyplot as plt

a = 5  # Lower limit
b = 10 # Upper limit, b > a

data = np.random.uniform(a, b, 3000)  # 3000 data generation
# print(data)
n, bins, patches = plt.hist(data, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
y = (1/(b-a))*np.ones(len(bins))
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold') 

titleText = "Uniform Distribution with a = " + str(a) + " and b = " + str(b) + "\n"      
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```

**GAUSSIAN DISTRIBUTION (CONTINUOUS DISTRIBUTION) - SPECIAL CASE : NORMAL DISTRIBUTION**

```python
# 4.  GAUSSIAN DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref: https://www.datacamp.com/community/tutorials/probability-distributions-python

import numpy as np
import matplotlib.pyplot as plt

# Special Case: Normal Distribution when mean = 0, sd = 1
mean = 10 
sd = 2

data = np.random.normal(mean, sd, 3000)  # 3000 data generation
# print(data)
n, bins, patches = plt.hist(data, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
y = 1/(sd * np.sqrt(2 * np.pi)) * np.exp( - (bins - mean)**2 / (2 * sd**2))
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Gaussian Distribution with $\mu$ = " + str(mean) + " and $\sigma$ = " + str(sd) + "\n"      
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```

**HALF-NORMAL DISTRIBUTION(CONTINUOUS DISTRIBUTION)**

```python
# 5.  HALF-NORMAL DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref : https://en.wikipedia.org/wiki/Half-normal_distribution

import numpy as np 
import matplotlib.pyplot as plt 
from scipy.stats import halfnorm

rv = halfnorm()
sd = 1
data = np.linspace(0, np.minimum(rv.dist.b, 3)) 
# print("Data : \n", data) 
y = np.sqrt(2)/(sd * np.sqrt(np.pi)) * np.exp( - (data**2) / (2 * sd**2) )
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(data, y, linewidth=3, color='r')
plt.plot(data, y_cdf, linewidth=3, color='k')
plt.plot(data, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Half-Normal Distribution with $\sigma$ = " + str(sd) + "\n"      
plt.title(titleText, fontweight ="bold")
plt.legend(LegendLabel, **LegendStyle)
plt.show()

# plot = plt.plot(distribution, rv.pdf(distribution)) 
```

**EXPONENTIAL DISTRIBUTION(CONTINUOUS DISTRIBUTION)**

```python
# 6.  EXPONENTIAL DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref: https://www.datacamp.com/community/tutorials/probability-distributions-python

import numpy as np
import matplotlib.pyplot as plt

lam = 0.5 # lambda value

data = np.random.exponential(1/lam , 3000)  # 3000 data generation
# print(data)
n, bins, patches = plt.hist(data, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
y = lam * np.exp( - lam*bins)
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Exponential Distribution with $\lambda$ = " + str(lam) + "\n"      
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```

**POISSON'S DISTRIBUTION(DISCRETE DISTRIBUTION)**

```python
# 7.  POISSON'S DISTRIBUTION (DISCRETE DISTRIBUTION)
# Ref: https://www.datacamp.com/community/tutorials/probability-distributions-python

import numpy as np
import matplotlib.pyplot as plt

def fact(n):  
    res = 1     
    for i in range(2, n+1): 
        res = res * i          
    return res 

n=80  # No of trials
p = 0.2 # probability 
mu = n*p # mean

result=[]
for i in range(1,n+1):
  y=(mu**i) * np.exp( - mu)/fact(i)
  result.append(y)
 
plt.plot(result,"or",result)
titleText = "Poisson Distribution with $\mu$ = " + str(mu) + "\n" 
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')
plt.title(titleText, fontweight ="bold") 
plt.show()
```

**GAMMA DISTRIBUTION(CONTINUOUS DISTRIBUTION)**

```python
# 8.  GAMMA DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref 1: https://www.datacamp.com/community/tutorials/probability-distributions-python
# Ref 2: https://en.wikipedia.org/wiki/Gamma_distribution
# Ref 3: https://mathworld.wolfram.com/GammaDistribution.html
# Ref 4: https://numpy.org/doc/stable/reference/random/generated/numpy.random.gamma.html

import numpy as np
import matplotlib.pyplot as plt
import math # for Gamma function
from scipy.special import gamma # for Gamma function

# Special Case: Normal Distribution when mean = 0, sd = 1
alpha = 2 # This is also denoted by k, known as shape parameter(always greater than 0) 
theta = 2 # This is also denoted by 1/beta, known as scale parameter (always greater than 0)
beta = 1/theta
data = np.random.gamma(alpha, theta, 3000)  # 3000 data generation
# print(data)
n, bins, patches = plt.hist(data, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
y = (beta**alpha)*(bins**(alpha-1))*np.exp(-beta*bins)/gamma(alpha) # gamma is imported from scipy
# y = (beta**alpha)*(bins**(alpha-1))*np.exp(-beta*bins)/(math.gamma(alpha)) # gamma is imported from math
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Gamma Distribution with k = " + str(alpha) + " and $\\theta$ = " + str(theta) + "\n"      
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```

**CHI-SQUARED DISTRIBUTION(CONTINUOUS DISTRIBUTION)**

```python
# 9.  CHI-SQUARED DISTRIBUTION (CONTINUOUS DISTRIBUTION)
# Ref 1: https://numpy.org/doc/stable/reference/random/generated/numpy.random.chisquare.html
# Ref 2: https://www.itl.nist.gov/div898/handbook/eda/section3/eda3666.htm

import numpy as np
import matplotlib.pyplot as plt
import math # for Gamma function

# Special Case: Normal Distribution when df = 1
df = 3 # degrees of freedom, also known as shape parameter, also denoted by k, it is integer and greater than 0

data = np.random.chisquare(df, 3000)  # 3000 data generation
# print(data)
n, bins, patches_ignored = plt.hist(data, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
y = (0.5**(df/2))*(bins**(df/2-1))*np.exp(-bins/2)/math.gamma(df/2)
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Chi-Squared Distribution with df = " + str(df) + "\n"      
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```

```python
# 10.  CHI-SQUARED DISTRIBUTION - Approach : SUM OF INDEPENDENT STANDARD NORMAL DISTRIBUTION
# Ref: https://en.wikipedia.org/wiki/Chi-square_distribution

import numpy as np
import matplotlib.pyplot as plt
import math


df = 3  # It indicates numbers of independent standard normal distribution(random variable) considered
data_len = 3000

x = [0]*data_len    # Initial all zero list
# type(x)
for i in range(df):
    x += (np.random.normal(0, 1, data_len))**2  # 3000 data generation
#     print(i)
# print(x)
type(x)
n, bins, patches = plt.hist(x, 100, density=True) # hist , data per bin = 100 + 1
# print(bins)
# print(n)
y = (0.5**(df/2))*(bins**(df/2-1))*np.exp(-bins/2)/math.gamma(df/2)
y_cdf = y.cumsum()
y_cdf /= y_cdf[-1]
survivalFunc=1-y_cdf
LegendLabel=["PDF, f(x)", "CDF, F(x)", "Survival Function, 1-F(x)"]
LegendStyle=dict(loc='upper left', bbox_to_anchor=(1.05, 1), fancybox=True, shadow=True)
plt.plot(bins, y, linewidth=3, color='r')
plt.plot(bins, y_cdf, linewidth=3, color='k')
plt.plot(bins, survivalFunc, linewidth=3, color='g')
plt.xlabel("x --->", fontweight='bold') 
plt.ylabel("f(x) / F(x) / 1-F(x) --->", fontweight='bold')

titleText = "Chi-Squared Distribution with df = " + str(df) + "\n"    
plt.title(titleText, fontweight ="bold") 
plt.legend(LegendLabel, **LegendStyle)
plt.show()
```
